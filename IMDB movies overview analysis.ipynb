### Exploratory Data Analysis

##### What kind of movies had highest budget and revenue?

```python
#Creating a function that return maximum value for certain column

def Maximum_Value(column):
    return df[df[column]==df[column].max()]
    
    
#starting with maximum budget 

Maximum_Value('budget')

```

```python
#For maximum revenue

Maximum_Value('revenue')
```

Movies from Action|Adventure|Fantasy|Science Fiction tend to have both high budget and revenue. Where by movie with highest budget was The Warrior's Way by Sngmoo Lee and highest revenue made by Avatar from James Cameron  



##### Is duration affect the popularity of movies?

```python
#For maximum runtime we can use our function 

Maximum_Value('runtime')
```

```python
#But for minimum runtime we cant use the function since we defined only for maximum outputs

df[df['runtime']==df['runtime'].min()]
```

```python
#Here we want to get movies with lowest runtime but highest popularity 

low_runtime=df[df['runtime']==df['runtime'].min()]

#highest popularity from lowest runtime
low_runtime[low_runtime['popularity']==low_runtime['popularity'].max()]
```

```python
#for highest popularite we can use our function 

Maximum_Value('popularity')
```



We can see that longest movies tend to have low popularity as from documentary The Story of Film: An Odyssey by Mark Cousins but short movies instead have highest popularity. But most popular movie is Jurassic World by Colin Trevorrow with duration of 124 minutes 


##### Who is the most famous director?

```python
#for highest popularite we can use our function 

Maximum_Value('popularity')
```

```python
#Here we want to count each appearance of directors in ascending order

df['director'].value_counts().sort_values(ascending=True)
```

### Function 

```python
def graph(a, b, dataframe, Title):
    plt.figure(figsize=(8,6))
    sns.boxplot(x= a, y= b, data=dataframe)
    plt.xticks(rotation=15)
    plt.title(Title);
```

```python
graph(df['original_title'][0:6], df['revenue'][0:6], df, 'Relationship between original title and revenue')
```

From these first 6 movies we can say Star Wars: The Force AwAkens had highest revenue

```python
sns.barplot(x =df['release_year'][0:5000], y =df['revenue'][0:5000], data = df, 
            palette ='plasma')
plt.title('Revenue Of different movies across years')
```

We can see 2003 had highest revenue projected near 200,000,000 dollars
